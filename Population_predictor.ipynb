## Selecting a desired model for population Prediction

```python
# defining a function so that we can use comparison between predited and actua population
def linePlot(output,label):
    plt.figure(figsize=(10, 6))
    plt.plot(label.values, label='True Values', color='blue', marker='o')
    plt.plot(output, label='Predicted Values', color='red', marker='x')
    plt.xlabel('Data points')
    plt.ylabel('Population')
    plt.title('Comparison of True Values and Predicted Values')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Splitting the data into Train data and Test data

Note: we have chosen the country as India for now because we must train all the model on the basis of population of that particular country of which we want to predict population and then test it.

```python
import numpy as np
from sklearn.model_selection import train_test_split

X = countries_pop['Year']
y = countries_pop['India']
#splitting the data set into training and testing set

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.20, random_state=42 )
X_train = X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
```

## 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
def linearRegression(data,label):
    model = LinearRegression()
    model.fit(X_train,Y_train)
    data = np.array(data).reshape(-1, 1)
    output = model.predict(data)

    print("Comparison between actual and predicted Population by Linear Regression")
    linePlot(output,label)    #will show the comparison between the predicted populatin and actual population

    return output

```

```python
# testing some trained data
some_data = X_train[:5]
some_label = Y_train[:5]

print(linearRegression(some_data,some_label))

print("labels")
print(list(some_label))
```

## 2. Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

def polynomialregression(degree,Yearin,label):             
    poly_features = PolynomialFeatures(degree=degree)          
    X_train_poly = poly_features.fit_transform(X_train) # transforms the existing features to higher degree features.
                 
    poly_model = LinearRegression() 
    poly_model.fit(X_train_poly, Y_train)    # fit the transformed features to Linear Regression    
    y_train_predicted = poly_model.predict(X_train_poly) # predicting on training data-set
    output = poly_model.predict(poly_features.fit_transform(Yearin))

    print("Comparison between actual and predicted Population by Polynomial Regression")
    linePlot(output,label)

    return output

```

```python
pred_pr = polynomialregression(3,some_data,some_label) #testing some trained data
print(pred_pr)
```

## 3. Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

def RandomForest(data,label):
    rfmodel = RandomForestRegressor()
    rfmodel.fit(X_train,Y_train)

    output = rfmodel.predict(data)

    print("Comparison between actual and predicted Population by Random Forest Regression")
    linePlot(output,label)

    return output

```

```python
predrf = RandomForest(some_data,some_label) #testing some trained data
print(predrf)
```

## Decision Tree Regression 

```python
from sklearn.tree import DecisionTreeRegressor

def DecisionTree(data,label):
    dtmodel = DecisionTreeRegressor(max_depth = 5)
    dtmodel.fit(X_train,Y_train)

    output = dtmodel.predict(data)

    print("Comparison between actual and predicted Population by Decision Regression")
    linePlot(output,label)

    return output

```

```python
preddt = DecisionTree(some_data,some_label) #testing some trained data
print(preddt)
```

## Evaluating the Model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
def rmse(true_value,pred_value):
    mse = mean_squared_error(true_value, pred_value)
    rmse = np.sqrt(mse)
    return rmse
```

```python
rmse_lr = rmse(Y_train,linearRegression(X_train,Y_train)) 
rmse_pr = rmse(Y_train,polynomialregression(3,X_train,Y_train))
rmse_rf = rmse(Y_train,RandomForest(X_train,Y_train))
rmse_dt = rmse(Y_train,DecisionTree(X_train,Y_train))

print("RMSE of Linear Regeression: ",rmse_lr)
print("RMSE of Polynomial Regression: ",rmse_pr)
print("RMSE of Random Forest Regression: ",rmse_rf)
print("RMSE of Decision Tree Regression: ",rmse_dt)
```
